```python
def convert(value):
    if value==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen']=df['SeniorCitizen'].apply(convert)
df.head()
```

```python


# Assuming `df` is your dataframe and contains columns 'SeniorCitizen' and 'Churn'

# Calculate percentages
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
total = grouped.groupby('SeniorCitizen')['Count'].transform('sum')
grouped['Percentage'] = grouped['Count'] / total * 100

# Pivot the data for a stacked bar chart
pivot = grouped.pivot(index='SeniorCitizen', columns='Churn', values='Percentage').fillna(0)

# Define a color palette with distinct colors
palette = sns.color_palette("viridis", n_colors=len(pivot.columns))

# Plot the stacked bar chart
fig, ax = mp.subplots(figsize=(6, 5))

bottom_values = [0] * len(pivot.index)  # Track bottom values for stacking

for idx, column in enumerate(pivot.columns):
    ax.bar(
        pivot.index, 
        pivot[column], 
        bottom=bottom_values, 
        color=palette[idx],  # Use a unique color for each category
        label=column
    )
    
    # Add percentage labels
    for i, val in enumerate(pivot[column]):
        if val > 0:  # Skip labels for 0% segments
            ax.text(
                i, 
                bottom_values[i] + val / 2, 
                f'{val:.1f}%', 
                ha='center', 
                va='center', 
                fontsize=9, 
                color='white'
            )
    # Update bottom values for the next stacked segment
    bottom_values = [b + v for b, v in zip(bottom_values, pivot[column])]

# Customize the plot
ax.set_title('Churn by SeniorCitizen (Percentage)')
ax.set_ylabel('Percentage')
ax.set_xlabel('SeniorCitizen')
ax.set_xticks(range(len(pivot.index)))
ax.set_xticklabels(pivot.index, rotation=0)  # Ensure labels are horizontal
ax.legend(title='Churn', loc='upper right')
mp.show()
```

```python

# Define the columns for which we want to create count plots
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Determine the layout of the subplots
n_cols = 3  # Number of plots per row
n_rows = -(-len(columns) // n_cols)  # Calculate rows required (ceiling division)

# Create the figure and axes
fig, axes = mp.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axes = axes.flatten()  # Flatten to easily iterate over axes

# Plot each column
for idx, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[idx], palette='viridis' , hue=df["Churn"])   # Explicitly set hue=None
    axes[idx].set_title(f'Count Plot for {col}')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Count')

# Remove any unused subplots
for idx in range(len(columns), len(axes)):
    fig.delaxes(axes[idx])  # Remove unused axes

# Adjust layout to avoid overlap
mp.tight_layout()
mp.show()
```

# Key observations show higher churn rates for customers using Fiber optic InternetService and those lacking value-added services like OnlineSecurity, DeviceProtection, and TechSupport. The KPIs include churn rate by service type and service adoption levels, highlighting retention gaps in premium services.
